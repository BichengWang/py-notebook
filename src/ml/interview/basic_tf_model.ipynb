```python
# Sequential

# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models

# Load dataset (here, MNIST for demonstration)
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the input data to be between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0

# Define a basic feed-forward neural network model
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # Flatten input from 2D to 1D
    layers.Dense(128, activation='relu'),  # First dense layer with 128 neurons and ReLU activation
    layers.Dropout(0.2),                   # Dropout layer to prevent overfitting
    layers.Dense(10, activation='softmax') # Output layer with 10 classes (softmax for multi-class classification)
])

# Compile the model with optimizer, loss function, and metrics
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model on training data
model.fit(x_train, y_train, epochs=5)

# Evaluate the model on test data
model.evaluate(x_test, y_test)
```
